### ONE HOT ENCODING With Keras Method

```python
import numpy as np

color_cats = np.unique(df['Renk Tercihi'].to_numpy())
occupation_cats = np.unique(df['Meslek'].to_numpy())
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
df["Renk Tercihi"] = le.fit_transform(df['Renk Tercihi'])
df["Meslek"] = le.fit_transform(df['Meslek'])

df
```

```python
from tensorflow.keras.utils import to_categorical
```

```python
ohe_color = to_categorical(df['Renk Tercihi'])
ohe_occupation = to_categorical(df['Meslek'])
```

```python
df.drop('Renk Tercihi', axis=1, inplace=True)
df[color_cats] = ohe_color
```

```python
df.drop('Meslek', axis=1, inplace=True)
df[occupation_cats] = ohe_occupation
```

### ONE HOT ENCODING With Unit Matrix Method



```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder

df = pd.read_csv('datasets/test_14.csv')

# print(df, end='\n\n')

color_cats = np.unique(df['Renk Tercihi'].to_numpy())
occupation_cats = np.unique(df['Meslek'].to_numpy())

le = LabelEncoder()
df['Renk Tercihi'] = le.fit_transform(df['Renk Tercihi'])
df['Meslek'] = le.fit_transform(df['Meslek'])

print(df, end='\n\n')

color_um = np.eye(len(color_cats))
occupation_um = np.eye(len(occupation_cats))

ohe_color = color_um[df['Renk Tercihi'].to_numpy()]
ohe_occupation = occupation_um[df['Meslek'].to_numpy()]

df.drop(['Renk Tercihi', 'Meslek'], axis=1, inplace=True)
df[color_cats] = ohe_color
df[occupation_cats] = ohe_occupation

print(df, end='\n\n')
```
