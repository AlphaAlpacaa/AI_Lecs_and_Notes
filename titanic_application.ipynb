```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('datasets/titanic_lec_14.csv')
df.head()
```

```python
df.drop(['PassengerId', 'Name', 'Ticket'], axis=1, inplace=True)
df.head(5)
```

```python
print("Toplam satır sayısı:", df.shape[0], "Toplam Sütun sayısı:", df.shape[1], end="\n\n")
```

```python
missing_info = df.isna().sum()
print("Eksik verilerin sütunlara göre dağılımı:")
print(missing_info, end="\n\n")
```

```python
missing_columns = [name for name in df.columns if df[name].isna().any()]
print("Eksik veri içeren sütunlar:")
print(missing_columns, end="\n\n")
```

```python
missing_ratio_per_column = {col: df[col].isna().sum().sum() / len(df) for col in missing_columns}
print("Eksik veri oranları:")
print(missing_ratio_per_column, end="\n\n")
```

```python
impute_age_value = df['Age'].mean().round()
impute_age_value
```

```python
df['Age'] = df['Age'].fillna(impute_age_value)
df
```

```python
impute_fare_value = df['Fare'].mean().round()
impute_fare_value
```

```python
df['Fare'] = df['Fare'].fillna(impute_fare_value)
df
```

```python
impute_cabin_value = "unknown"
df['Cabin'] = df['Cabin'].fillna(impute_cabin_value)
df
```

```python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()
```

```python
df['Sex'] = label_encoder.fit_transform(df['Sex'])
df.head()
```

```python
label_encoder.inverse_transform(df['Sex'])
```

```python
from sklearn.preprocessing import OneHotEncoder

ohe = OneHotEncoder(sparse_output=False)
```

```python
ohe.fit(df[['Pclass']])
```

```python
ohe_data = ohe.transform(df[['Pclass']])
ohe_data
```

```python
df.drop('Pclass', axis=1, inplace=True)
df.head()
```

```python
df[['Pclass1', 'Pclass2', 'Pclass3']] = ohe_data
df
```
