```python
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.activations import relu

import pandas as pd 
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('datasets/heart_failure_clinical_records_dataset.csv')

df.head(5)
```

```python
df.describe()
```

```python
dataset_x = df.iloc[:, :-1]
dataset_y = df.iloc[:, -1]
```

```python
training_dataset_x, testing_dataset_x, training_dataset_y, testing_dataset_y = train_test_split(dataset_x, dataset_y, test_size=0.1)
len(training_dataset_x), len(training_dataset_y), len(testing_dataset_x), len(testing_dataset_y)
```

```python
model = Sequential(name="Health_Failure_Checker")

model.add(Dense(32, activation=relu, input_shape=(dataset_x.shape[1],), name="Hidden-1"))
model.add(Dense(32, activation=relu, name="Hidden-2"))
model.add(Dense(1, activation='sigmoid', name="Output"))

model.summary()
```

```python
from tensorflow.keras.optimizers import RMSprop

rmsprop = RMSprop()
model.compile(optimizer=rmsprop, loss='binary_crossentropy', metrics=['binary_accuracy'])
```

```python
model.fit(training_dataset_x, training_dataset_y, epochs=200, batch_size=32, validation_split=0.1)
```
